```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
for i in train.columns:
    if train[i].dtype == 'O':
        train[i] = lb_make.fit_transform(train[i])
        test[i] = lb_make.fit_transform(test[i])
```

```python
## features to be included in the model
feature_names = [x for x in train.columns if x not in ['s.no','Timestamp',
                                                      'comments', 'treatment']]
target = train['treatment']
```

```python
x_train, x_test, y_train, y_test = train_test_split(train[feature_names].values, target,
                                                    test_size=0.2,
                                                    random_state=2019)
model = xgb.XGBClassifier()
model.fit(x_train, y_train)
print(model.feature_importances_)
```

```python
fea_imp_ = pd.DataFrame({'cols':train[feature_names].columns, 'fea_imp':model.feature_importances_})
fea_imp_.loc[fea_imp_.fea_imp > 0].sort_values(by=['fea_imp'], ascending = False).reset_index()
```

```python
# feature selection based on important score
feature_names = list(fea_imp_[fea_imp_['fea_imp']>0.01]['cols'])
feature_names
```

```python
from sklearn.metrics import roc_auc_score,classification_report,accuracy_score
from sklearn.metrics import roc_curve
from sklearn.neighbors import KNeighborsClassifier
f = []
for i in feature_names:
    f.append(i)
    print(f)
    x_train, x_test, y_train, y_test = train_test_split(train[f].values, target,
                                                        test_size=0.2,
                                                        random_state=2019)
    model = xgb.XGBClassifier()
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print("Accuracy: ",accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    # AUC ROC CURVE
    predictions = model.predict_proba(x_test)[:,1]
    roc_auc_score(y_test, y_pred, average='weighted')

    fpr, tpr, threshold = roc_curve(y_test, predictions)

    # Calculating the AUC Score
    auc = np.trapz(tpr,fpr)
    pltTitle = print("AUROC Plot:", "%.4f" %auc)

    # Plotting the ROC Curve
    plt.plot(fpr,tpr)
    plt.title(pltTitle)
    plt.show()
```
